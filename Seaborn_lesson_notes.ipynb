```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns  #use sns for any time we access sns datasets.  sns load_dataset
# the above is basically the default import for our use
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips)
#relplot produces a scatter plot by default, but can be changed with kind=___
#x gives us the data for the x axis, y gives us the data for the y axis, hue shows different colors for info in column
#data tells us what dataset to use
```

```python
sns.relplot(x='total_bill', y='tip', col='smoker', data=tips)
#col is giving us subplots (column plots) for the info in the 'smoker' column
```

```python
sns.relplot(x='total_bill', y='tip', style='smoker', data=tips)
#notice the style gives us different shapes for the info in the 'smoker' column
```

```python
sns.relplot(x='total_bill', y='tip', hue='size', data=tips)
#the column size in the tips dataset has 6 different unique values, so there are 6 colors.  
#Seaborn defaults the color choice for this.
```

```python
## if you don't want to see a scatterplot bc it doesn't make sense for the data you're showing
# use histplot

sns.histplot(tips.total_bill)
plt.show()
```

```python
#if you want an approximation for a specific bill, 
#kde=True it will show a line that is giving an estimation.
sns.histplot(tips.total_bill, kde=True)
plt.show()
```

```python
#boxplot sns.boxplot  it only needs an x value, but can take an x and y
#suptitle will title the subplot. plt.suptitle
#plt.title will give the entire grouping a title

sns.boxplot(data=tips, x="tip")
plt.show()
```

```python
sns.boxplot(data=tips, x="tip", y='time', hue='smoker')
plt.show()
```

```python
sns.catplot(data=tips, y='day', x='tip', hue='smoker', col='time')
plt.show()
```

```python
#heatmaps are great for crosstabs
data = pd.crosstab(tips.time, tips['size'], normalize=True)
data
```

```python
sns.heatmap(data, annot=True, cmap=plt.cm.Greens)
```

```python
sns.pairplot(tips)  #gives an at-a-glance visual of every numeric pairing of data from the dataset.
```
