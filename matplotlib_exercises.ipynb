```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import math
```

```python
x = pd.Series(range(-10, 20)) # or some other range of numbers for your x values

```

```python
def squared_minus_x_plus_two(x):
    
    return x ** 2 -x + 2

y = x.apply(squared_minus_x_plus_two)
```

```python
plt.annotate('Origin', xy=(0, 0), xytext=(-4, 50),
             arrowprops={'facecolor': 'blue'})

plt.plot(x, y)


plt.show()
```

```python
y = [(n ** 3) for n in x]  #my formula is now y = x**3. I already have a range for x set up above.
```

```python
plt.plot(x,y)  #this plots the points on a graph
plt.show()     #this shows the graph
```

```python
y = [(2 ** n) for n in x]  #my formula is now y = 2**x.  I'm still using the x values above.
```

```python
plt.plot(x,y)  #this plots the points on a graph
plt.show()     #this shows the graph
```
