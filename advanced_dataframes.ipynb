```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

### What is the result of using a right join on the DataFrames?

### What is the result of using an outer join on the DataFrames?

```python
# right join:
#default concat first:

pd.concat([users, roles])
```

```python
join_right = users.merge(roles, left_on='role_id', right_on='id', how='right', indicator=True)

join_right
```

```python
# outer join: 

join_outer = users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)

join_outer
```

### What happens if you drop the foreign keys from the DataFrames and try to merge them?

```python
no_foreign_keys = (users.merge(roles, left_on='role_id', right_on='id', how='right', indicator=True)
                        .drop(columns = 'role_id')
                        .rename(columns={'id_x': 'id', 
                                         'name_x': 'employee',
                                         'id_y': 'role_id',
                                         'name_y': 'role'}
                                            ))

no_foreign_keys
```

### Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.

```python
mpg['mileage_difference'] = (mpg.hwy - mpg.cty)

mpg.head()
```

### Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.

```python
mpg["average_mileage"] = ((mpg.hwy + mpg.cty)/2)

mpg.head()
```

### Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

```python
mpg["is_automatic"] = np.where(mpg.trans.str.contains('auto'), True, False)

# df['passing_math'] = np.where(df.math < 70, 'failing', 'passing')
# df

mpg.head()
```

### Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?

```python
#group by manufacturer - average mileage

# Here's an example from the lesson: df.groupby('classroom').math.agg(['min', 'mean', 'max'])
# Let's apply it here:

mileage_by_manufacturer = mpg.groupby('manufacturer').average_mileage.agg(['mean'])

mileage_by_manufacturer
```

```python
# Now, let's sort the mileage column:

mileage_by_manufacturer.sort_values(by = 'mean', ascending = False)
```

```python
#we can chain the method for the top now to get the max

mileage_by_manufacturer.sort_values(by = 'mean', ascending = False).head(1)
```
